```python
import pandas as pd

def summarize(n):
    #files are like data-1.csv
    # data-   i    .csv
    filename = "data-" + str(n) + ".csv"
    df = pd.read_csv(filename)

    print("data-" + str(n) + " (n=" + str(len(df)) + ")")

    #summary for x and y
    for col in ["x", "y"]:
        values = df[col]

        mean_value = values.mean()
        std_value = values.std()
        min_value = values.min()
        max_value = values.max()

        print("  " + col + ": mean=" + str(mean_value) +
              "  std=" + str(std_value) +
              "  min=" + str(min_value) +
              "  max=" + str(max_value))

    # correlation
    r = df["x"].corr(df["y"])
    print("  correlation r = " + str(r))

    return df


data = []

for n in range(1, 7):
    result = summarize(n)
    data.append(result)
```
